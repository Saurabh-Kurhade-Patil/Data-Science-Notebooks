```python
class myLinearRegression:
    def __init__(self):
        self.w = None
        self.b = None

    def fit(self, X_train, y_train):
        numerator = 0
        denominator = 0
        for i in range(X_train.shape[0]):
            numerator = numerator + ((X_train[i]-X_train.mean())*(y_train[i]-y_train.mean())) # ∑(xi−x̄)(yi−ȳ)
            denominator = denominator + ((X_train[i]-X_train.mean())**2) # ∑(xi−x̄)**2

        self.w = numerator/denominator # m=∑(xi−x̄)(yi−ȳ)/∑(xi−x̄)**2
        self.b = y_train.mean() - (self.w*X_train.mean()) # b = y -wX

        print(f'values for w is {self.w} and b is {self.b}.')
        return (self.w ,self.b)


    def predict(self, X_test):
        return (self.w * X_test.mean()) + self.b # y = wX +b
    
```

```python
lr = myLinearRegression() 
```
